<h3>Iteration over product links id db and extract product information from each product link</h3>
<p>Important note! products_all is an extended table that contains columns with product features like name,price etc.</p>

```python
%%time
db_string = 'postgresql://postgres:test1234@localhost:5432/oma_catalog_test'
db = create_engine(db_string)
update_set = db.execute("SELECT * FROM products_all WHERE is_parsed IS NULL LIMIT 500") 
meta = MetaData(db) 
products_all = Table('products_all', \
                               meta, autoload=True, autoload_with=db)

for entry in update_set: 
    response = simple_get(entry.url)
    print(f'entry.id is {entry.id}')
    if response is not None:
        soup = BSoup(response, 'html.parser')
        p_dict = get_product_parameters(soup)
    with db.connect() as conn:
        dt = datetime.now()
        if entry.is_parsed == None:
            update_sttm = products_all.update().where(products_all.c.id==entry.id).values(product_name = p_dict['product_name'], \
                                                                                                   product_price = float(p_dict['product_price']), \
                                                                                                   product_description = p_dict['product_description'], \
                                                                                                   product_characteristics = p_dict['product_characteristics'], \
                                                                                                   similar_products = p_dict['similar_products'], \
                                                                                                   product_image_link = p_dict['product_image_link'], \
                                                                                                   product_is_hit = p_dict['product_is_hit'], \
                                                                                                   is_parsed = dt)                                                                                                    
            conn.execute(update_sttm)
            print('parsed')
        else:
            pass
```
